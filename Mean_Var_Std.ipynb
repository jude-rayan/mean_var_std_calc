```python
### #import mean_var_std
import numpy as np
import pprint
from unittest import main

result = {"mean":[],"variance":[],"standard_deviation":[],"max":[],"min":[],"sum":[]};

def means(arr):    
    mean_array = []
    mean_array.append(np.mean(arr,axis=0).tolist())
    mean_array.append(np.mean(arr,axis=1).tolist()) 
    mean_array.append(np.mean(arr.flatten())) 
    result["mean"] = mean_array
    return None
    
def variances(arr):
    variance_array = []
    variance_array.append(np.var(arr,axis=0).tolist())
    variance_array.append(np.var(arr,axis=1).tolist())
    variance_array.append(np.var(arr.flatten()))   
    result["variance"] = variance_array
    return None
    

def std_devs(arr):
    stdv_array = []
    stdv_array.append(np.std(arr,axis=0).tolist())
    stdv_array.append(np.std(arr,axis=1).tolist())
    stdv_array.append(np.std(arr.flatten())) 
    result["standard_deviation"] = stdv_array
    return None

def max_calc(arr):
    max_array = []
    max_array.append(np.max(arr,axis=0).tolist())
    max_array.append(np.max(arr,axis=1).tolist())
    max_array.append(np.max(arr.flatten())) 
    result["max"] = max_array
    return None


def min_calc(arr):
    min_array = []
    min_array.append(np.min(arr,axis=0).tolist())
    min_array.append(np.min(arr,axis=1).tolist())
    min_array.append(np.min(arr.flatten())) 
    result["min"] = min_array
    return None
    
def calc_sum(arr):
    sum_array = []
    sum_array.append(np.sum(arr,axis=0).tolist())
    sum_array.append(np.sum(arr,axis=1).tolist())
    sum_array.append(np.sum(arr.flatten())) 
    result["sum"] = sum_array
    return None
    

def calculate(list):
    if len(list)<9:
        raise ValueError("List must contain nine numbers.")
    else:   
      input_array = np.array(list).reshape(3,3)
      means(input_array)
      variances(input_array)
      std_devs(input_array)
      max_calc(input_array)
      min_calc(input_array)
      calc_sum(input_array)
    return result
    
  

  
#result["mean"].append([12,5])
  

calculate([9,1,5,3,3,3,2,9,0])
```
